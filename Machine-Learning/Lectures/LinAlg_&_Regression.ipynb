# Linear Algebra Approach

#Y represents the dependent variable (life satisfaction),X is the independent variable (GDP per capita).
#β0 is the intercept, β1 is the slope of the line, and ϵ is the error term.

#Linear Regression Modeling

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Given data
data = {
    'Country': ['Hungary', 'Korea', 'France', 'Australia', 'United States'],
    'GDP_per_Capita': [12240, 27195, 37675, 50962, 55805],
    'Life_Satisfaction': [4.9, 5.8, 6.5, 7.3, 7.2]
}

# Create DataFrame
df = pd.DataFrame(data)

# Linear regression calculation
X = np.vstack([np.ones(len(df)), df['GDP_per_Capita']]).T
Y = df['Life_Satisfaction'].values.reshape(-1, 1)

# Calculate coefficients
beta_hat = np.linalg.inv(X.T @ X) @ X.T @ Y

# Scatter plot of the data
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP_per_Capita'], df['Life_Satisfaction'], color='blue', label='Data points')

# Regression line
x_values = np.array([min(df['GDP_per_Capita']), max(df['GDP_per_Capita'])])
y_values = beta_hat[0] + beta_hat[1] * x_values
plt.plot(x_values, y_values, color='red', label='Regression line')

# Annotating the plot with the regression equation
equation_text = f"Life Satisfaction = {beta_hat[0][0]:.2f} + {beta_hat[1][0]:.5f} * GDP per Capita"
plt.text(min(df['GDP_per_Capita']), min(df['Life_Satisfaction']), equation_text, fontsize=12)

# Plot details
plt.xlabel('GDP per Capita')
plt.ylabel('Life Satisfaction')
plt.title('Life Satisfaction vs GDP per Capita')
plt.legend()
plt.grid(True)
plt.show()
```
